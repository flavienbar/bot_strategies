```python
import pandas as pd
import datetime as dt
from dateutil.relativedelta import relativedelta
import numpy as np
```

```python
btc_raw_df = pd.read_parquet("/home/giujorge/Documents/Projets/bot_strategies/data/raw/BTC-USDT.csv.parquet.gzip")
btc_raw_df
```

# Calculate montly returns by cumulating minutes returns

```python
 df = btc_raw_df[['close']].pct_change().resample('M').agg(lambda x:(x+1).prod() -1).copy()
 df.head()
```

# Calculate returns over the past 11 months

```python
pass_11 = (df +1).rolling(11).apply(np.prod)-1
pass_11.head(20)
```

```python
pass_11.tail(20)
```

```python
formation = dt.datetime(2022, 3, 31)
formation
```

```python
end_mesurement  = formation - relativedelta(months=1)
end_mesurement
```

```python
ret_12 = pass_11.loc[end_mesurement]
ret_12.reset_index()
```

```python
# besoin de comparer plus de 
```

```python
# Add a column containing the quitiles
```

```python
ret_12['quintile'] = pd.qcut(ret_12.iloc[:,1], 5, labels=False)
```

```python
# Get winers and losers return by finding the last months return of winner /loser symbols
```

```python
winners = ret_12[ret_12.quantile == 4]
winners
```

```python
losers = ret_12[ret_12.quantile == 0]
losers
```

```python
winnersret = df.loc[formation + relativedelta(months=1), df.columns.isin(winners.Symbols)]
winnersret
```

```python
losersret = df.loc[formation + relativedelta(months=1), df.columns.isin(losers.Symbols)]
losersret
```

```python
# Calculate profit of mumentum strategy
```

```python
momentumprofit = winnersret.mean() - losersret.mean()

momentumprofit
```
